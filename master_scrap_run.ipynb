```python
import pandas as pd
import numpy as np
import os
import time
import warnings
warnings.filterwarnings('ignore')
import time
from datetime import datetime, timedelta
import nltk
nltk.download('punkt')
todaysdate=datetime.now().strftime("%Y-%m-%d")
last_date=(datetime.now() + timedelta(days=-6)).strftime("%Y-%m-%d")
import logging
```

```python
# from Theedgemarkets_news_1 import  the_edge_scarp
from Theedgemarkets_news_2 import  the_edge_scarp
from CNN_1 import  cnn_scrap
from reuters_1 import  reuters_scrap
from The_star_1 import  the_star_scrap
from Bloomberg_1 import  bloomberg_scrap
keyword=['RP TRADING','CIMB bank','RHB bank','maybank']

logging.basicConfig(level=logging.DEBUG,filename='Execution.log',filemode='w',format='%(asctime)s - %(levelname)s - %(message)s')
with open('Execution.log', 'w'):
    pass
logging.debug('********************Execution started *******************')
logging.debug('Libraries imported successfully')
```

```python
try:
    df_the_edge=the_edge_scarp(keyword)
    df_the_edge['trigger']=np.where(((df_the_edge['date_time']>=last_date)&(df_the_edge['date_time']<=todaysdate))
                                & (df_the_edge['Match_1']>90),'Y','N')
    df_the_edge=df_the_edge[['keyword', 'source','url','date_time','title','article','Match_1','trigger']].copy()
    df_the_edge.columns=['Keyword','Source','URL','Date','Headline','News_Article','Match','Trigger']
    logging.debug("****************Successfully scrap The Edge market news website****************")
except Exception as ex:
    logging.exception('Error while scraping The Edge market news website ', ex)
```

```python
# try:
#     df_the_edge=the_edge_scarp(keyword)
#     df_the_edge['trigger']=np.where(((df_the_edge['date_time']>=last_date)&(df_the_edge['date_time']<=todaysdate))
#                                 & (df_the_edge['Match_1']>90),'Y','N')
#     df_the_edge=df_the_edge[['keyword', 'source','url','date_time','title','article','Match_1','trigger']].copy()
#     df_the_edge.columns=['Keyword','Source','URL','Date','Headline','News_Article','Match','Trigger']
#     print("****************Successfully scrap The Edge market news website****************")
# except Exception as ex:
#     print('Error while scraping The Edge market news website ', ex)
try:
    cnn_df=cnn_scrap(keyword)
    cnn_df['date_time'] = cnn_df["Date"].apply(lambda x : datetime.strptime(str(x), '%b %d, %Y') if x!='' else None)
    cnn_df['trigger']=np.where(((cnn_df['date_time']>=last_date)&(cnn_df['date_time']<=todaysdate))
                                & (cnn_df['Match_1']>90),'Y','N')
    cnn_df=cnn_df[['keyword', 'website','headline_url','date_time','Headline','article','Match_1','trigger']].copy()
    cnn_df.columns=['Keyword','Source','URL','Date','Headline','News_Article','Match','Trigger']
    logging.debug("****************Successfully scrap CNN news website****************")
except Exception as ex:
    logging.exception('Error while scraping CNN news website ', ex)   
    
    
try:
    df_reuters=reuters_scrap(keyword)
    df_reuters['trigger']=np.where(((df_reuters['date_time']>=last_date)&(df_reuters['date_time']<=todaysdate))
                                & (df_reuters['Match_1']>90),'Y','N')
    df_reuters=df_reuters[['keyword', 'source','url','date_time','title','article','Match_1','trigger']].copy()
    df_reuters.columns=['Keyword','Source','URL','Date','Headline','News_Article','Match','Trigger']
    logging.debug("****************Successfully scrap Reuters news website****************")
except Exception as ex:
    logging.exception('Error while scraping Reuters news website ', ex)  

try:
    df_the_star=the_star_scrap(keyword)
    df_the_star['trigger']=np.where(((df_the_star['date_time']>=last_date)&(df_the_star['date_time']<=todaysdate))
                                & (df_the_star['Match_1']>90),'Y','N')
    df_the_star=df_the_star[['keyword', 'source','url','date_time','title','article','Match_1','trigger']].copy()
    df_the_star.columns=['Keyword','Source','URL','Date','Headline','News_Article','Match','Trigger']
    logging.debug("****************Successfully scrap The Star news website****************")
except Exception as ex:
    logging.exception('Error while scraping THe Star news website ', ex)      
    
    
try:
    df_Bloomberg=bloomberg_scrap(keyword)
    def date_con(x):
        if x[0].isalpha():
            a=datetime.strptime(str(" ".join(str(x).split(",")[0:2]).replace(" ",",").replace(",,",",")), '%B,%d,%Y')
        else:
            a=None
        return a        
    df_Bloomberg['date_time'] = df_Bloomberg["Date"].apply(lambda x: date_con(x))
    df_Bloomberg['trigger']=np.where(((df_Bloomberg['date_time']>=last_date)&(df_Bloomberg['date_time']<=todaysdate))
                            & (df_Bloomberg['Match_1']>90),'Y','N')
    df_Bloomberg=df_Bloomberg[['Keyword', 'Source','headline_url','date_time','Headline','Article','Match_1','trigger']].copy()
    df_Bloomberg.columns=['Keyword','Source','URL','Date','Headline','News_Article','Match','Trigger']
    logging.debug("****************Successfully scrap Bloomberg news website****************")
except Exception as ex:
    logging.exception('Error while scraping Bloomberg news website ', ex) 
    
logging.shutdown()       
```
