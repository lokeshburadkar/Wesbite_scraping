```python
import webbrowser
import requests
from bs4 import BeautifulSoup
import os
import time
import re
from selenium import webdriver
chrome_path=r"C:\Users\ADMIN\Desktop\CIMB Work\chromedriver.exe"
from datetime import datetime, timedelta
todaysdate=datetime.now().strftime("%Y-%m-%d")
last_date=(datetime.now() + timedelta(days=-6)).strftime("%Y-%m-%d")
keyword=['bank','RP TRADING','cimb bank','rhb bank']
url_base=[]
headline=[]
headline_url=[]
cat=[]
for key in keyword:
    base_url="https://www.thestar.com.my/search/?q={}&qguid=&qtag=&QDR=QDR_simple&DateRange_DDL=day&qsort=newest&qrec=30".format(key)
    print(base_url)
    driver=webdriver.Chrome(chrome_path)
    driver.get(base_url)
    time.sleep(3)
    real_soup = BeautifulSoup(driver.page_source, 'html.parser')
    driver.quit()
    ab = real_soup.find_all('div',class_="row list-listing") 
    if(ab!=None):
        for i in ab:
            anchor = i.find('h2', {'class':'f18'}).find('a')
            link = anchor['href']
            print(link)
            headline_url.append(link)
            headline.append(anchor['data-content-title'])
            print(anchor['data-content-title'])
            url_base.append(base_url)
            cat.append(key)
    else:
        print("No result found")
        headline_url.append("No url found")
        headline.append("No headline found")
        url_base.append(base_url)
        cat.append(key)
    print("*"*20,"END","*"*20)  
```

```python
import pandas as pd
df = pd.DataFrame({'keyword':cat,'main_url':url_base, 'Headline':headline,'headline_url': headline_url })
df.drop_duplicates(inplace=True)
df.shape  
```

```python
import nltk
nltk.download('punkt')
import webbrowser
import re
import time
from newspaper import Article
from newspaper import Config
user_agent = 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10.15; rv:78.0) Gecko/20100101 Firefox/78.0'
config = Config()
config.browser_user_agent = user_agent
title=[]
summary=[]
article=[]
key=[]
date_time=[]
url_1=[]
for url in df.headline_url:
    if url!='No url found':
        print(url)
        time.sleep(5)
        url_1.append(url)
        #For different language newspaper refer above table
        toi_article = Article(url, language="en",config=config) # en for English
        #To download the article
        toi_article.download()
        #To parse the article
        toi_article.parse()
        #To perform natural language processing ie..nlp
        toi_article.nlp()
        #To extract title
        print("Article's Title:")
        print(toi_article.title)
        title.append(toi_article.title)
        print("n")
        #To extract text
        print("Article's Text:")
        print(toi_article.text)
        article.append(toi_article.text)
        print("n")
        #To extract summary
        print("Article's Summary:")
        print(toi_article.summary)
        summary.append(toi_article.summary)
        print("n")
        #To extract keywords
        print("Article's Keywords:")
        print(toi_article.keywords)
        key.append(toi_article.keywords)
        date_time.append(toi_article.publish_date)
        
    else:
        url_1.append(url)
        title.append("")
        article.append("")
        summary.append("")
        key.append("")
        date_time.append("")
```
