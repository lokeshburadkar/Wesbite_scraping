```python
import requests
from bs4 import BeautifulSoup
import os
import time
from datetime import datetime, timedelta
todaysdate=datetime.now().strftime("%Y%m%d")
last_date=(datetime.now() + timedelta(days=-6)).strftime("%Y%m%d")
# keyword=['Fraud', 'Laundering', 'Bribe', 'Corruption', 'Embezzlement', 'Organized crime',
#          'Scam', 'Smuggling', 'Theft', 'Tax evasion', 'Insider trading', 'Ransom',
#          'Illegal gambling', 'Forgery', 'Counterfeit', 'Cybercrime', 'Murder']
# # keyword=['Murder']
keyword=['RP TRADING','cimb bank','rhb bank']
# keyword=['Fraud', 'Laundering','Insider trading','Murder']
url_list=[]
headline_link=[]
url_search=[]
category=[]
page_url_list=[]
for key in keyword:
    base_url="https://www.thesundaily.my/search-result/-/search/{}/false/true/{}/{}/date/true/true/0/0/meta/0/0/0/1".format(key,last_date,todaysdate)
    print(base_url)
    r1 = requests.get(base_url)
    time.sleep(6)
    coverpage = r1.text
    soup1 = BeautifulSoup(coverpage,"html.parser")
    if (soup1.find('div',{"id":"_2226326732_mylistCarrousel"})!=None):
        weblinks = soup1.find('div',{"id":"_2226326732_mylistCarrousel"})
        ab=weblinks.find_all("ul",{"id":"_2226326732_myNavButtons"})
        print("*"*50)
        for i in ab:
            abc=i.text.strip()
            page_no=abc.split()
            for page in page_no:
                break_url=base_url.rsplit('/', 1)[0]
                final_url=break_url+"/"+page
                print(final_url)
                r1 = requests.get(final_url)
                coverpage = r1.text
                soup1 = BeautifulSoup(coverpage,"html.parser")
                if (soup1.find('ul',class_="noticias")!=None): 
                    
                    weblinks = soup1.find('ul',class_="noticias")
                    # weblinks
                    ab=weblinks.find_all("a",href=True)
                    append_url="https://www.thesundaily.my"
#                     head=[]
                    for sublink in ab:
                        url_link=sublink.get("href")
                        url_list.append(append_url+url_link)
                        print(append_url+url_link)
                        headline=str(sublink.text.strip())
#                         head.append()
                        headline_link.append(headline)
                        print(headline)
                        category.append(key)
                        # add base url with page no
                        page_url_list.append(final_url)  
                else:
                    print("no data present under link")
                    headline_link.append("NO Headline")
                    url_list.append("No Headline Link")
                    category.append(key)
                    page_url_list.append(base_url)      
                
        print("*"*50)
    else:
        print("No pages present")
        r1 = requests.get(base_url)
        coverpage = r1.text
        soup1 = BeautifulSoup(coverpage,"html.parser")
        
        if (soup1.find('ul',class_="noticias")!=None):        
            weblinks = soup1.find('ul',class_="noticias")
            ab=weblinks.find_all("a",href=True)
            append_url="https://www.thesundaily.my"
            for sublink in ab:
                url_link=sublink.get("href")
                url_list.append(append_url+url_link)
                print(append_url+url_link)
                headline=str(sublink.text.strip())
                headline_link.append(headline)
                print(headline)
                # add category for each url
                category.append(key)
                # add base url with page no
                page_url_list.append(base_url)
  
        else:
            print("no data present under link")
            headline_link.append("NO Headline")
            url_list.append("No Headline Link")
            category.append(key)
            page_url_list.append(base_url)

    print("#"*20,"next url")
    ## final url
    url_search.append(base_url)
 
```

```python
#### creating dataframe     
import pandas as pd
df = pd.DataFrame({'keyword':category,'main_url':page_url_list, 'Headline':headline_link,'url': url_list })
df.drop_duplicates(inplace=True)
df.shape  
```

```python
import nltk
nltk.download('punkt')
import webbrowser
import re
import time
from newspaper import Article
from newspaper import Config
user_agent = 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10.15; rv:78.0) Gecko/20100101 Firefox/78.0'
config = Config()
config.browser_user_agent = user_agent
title=[]
summary=[]
article=[]
key=[]
date_time=[]
url_1=[]
for url in df.url:
    if url!='No Headline Link':
        print(url)
        url_1.append(url)
        #For different language newspaper refer above table
        toi_article = Article(url, language="en",config=config) # en for English
        #To download the article
        toi_article.download()
        #To parse the article
        toi_article.parse()
        #To perform natural language processing ie..nlp
        toi_article.nlp()
        #To extract title
        print("Article's Title:")
        print(toi_article.title)
        title.append(toi_article.title)
        print("n")
        #To extract text
        print("Article's Text:")
        print(toi_article.text)
        article.append(toi_article.text)
        print("n")
        #To extract summary
        print("Article's Summary:")
        print(toi_article.summary)
        summary.append(toi_article.summary)
        print("n")
        #To extract keywords
        print("Article's Keywords:")
        print(toi_article.keywords)
        key.append(toi_article.keywords)
        date_time.append(toi_article.publish_date)
        
    else:
        url_1.append(url)
        title.append("")
        article.append("")
        summary.append("")
        key.append("")
        date_time.append("")
```

```python
df["article"]=article
df["date_time"]=date_time
df["title"]=title
df["summary"]=summary
df["key"]=key
df["url"]=url_1
df["source"]="thesundaily"
```

```python
import re
import time
date_time_list=[]
article_summary=[]
for i in df.url:
    if i!="No Headline Link":
        print(i)
        r1 = requests.get(i)
        time.sleep(6)
        coverpage = r1.text
        soup1 = BeautifulSoup(coverpage,"html.parser")
        weblinks = soup1.find('div',{"class":"paragraph"})
        ab=weblinks.find_all("p")
        date=soup1.find('div',class_="datefrom small")
        date_time=date.text.strip().replace("/","").replace("H.","").strip()
        date_time_list.append(date_time)
        print(date_time)
        summary=[]
        for i in ab:
            article=str(re.sub(r'\s+',' ',i.text.strip()))
            print(str(re.sub(r'\s+',' ',i.text.strip())))
            summary.append(article)
            fn_out=' '.join([str(i) for i in summary])
        article_summary.append(fn_out)
    else:
        print("No Headline URL present")
        date_time_list.append("NO URL PRESENT")
        article_summary.append("NO URL PRESENT")
        
### adding both article and date time in df    
df["Article"]=article_summary
df["Date_Time"]=date_time_list    
```
